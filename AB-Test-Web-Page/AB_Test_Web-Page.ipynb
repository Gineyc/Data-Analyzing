# 3. A/B Test
Based on the analysis above, we can get the conversion rate of treatment group and control group is at 11.892% and 12.04%, respectively.
The difference of probability between control and treatment groups is tiny. Therefore, we still need more evidence to assess the performance of treatment group.There is a need to conduct a hypothesis test.

## Hypotheses Test: 

 P new - P old represents the diffrences in converted success rate between the old and new pages. 

- Null Hypothesis: There is no difference between the control and experiment group.
   -  P new - P old = 0
  
- Alternate Hypothesis: There is difference between the control and experiment group.
   -   P new - P old != 0

In the following analysis, we need to perform the sampling distrubention of the difference in convertion rate between the control group and treatment group over 10,000 iterations, to calculate an estimate from the null.

```python
print("size of control group: ",control.count()['user_id'])
print("size of treatment group: ",treatment.count()['user_id'])
n_new = treatment.count()['user_id']
n_old = control.count()['user_id']
```

```python
# define a function for permutation
def perm_fun(n,treament_size,control_size):
    # conduct a in-place shuffle the population of converted
    np.random.shuffle(n) 
    #resample according to the size of each group
    n1=n[:treament_size]
    n2=n[-control_size:]
    #calculate P new - P old
    return n1.mean()-n2.mean()

#Number of iterations
num=10000
diff=[]
n=np.array(df_c.converted)
for i in range(num):
    diff.append(perm_fun(n,n_new,n_old))
#Observed difference
obs_diff=treatment.converted.mean()-control.converted.mean()

plt.axvline(x=obs_diff,c='r')
sns.distplot(diff)
```

```python
print("The propotion of observed difference > propotion of difference",round((diff>obs_diff).mean()*100,2),"%")
```

## Sampling distribution analysis

- The proportion of the conversion rate differences was greater than the observed difference (88.65%). The p-value is way larger than the type error I (Mistakenly concluding an effect is real (when it is due to chance) rate 5%. It means we fail to rejecet the null hypothesis.

- P-value is the probability of obtaining results as unusual or extreme as the observed result. When the null hypothesis is true, the statistical summary would be the same as or of greater magnitude than the actually observed results.

- When the p-value is low, it suggests that the null hypothesis is not true, and we need to consider the alternatice hypothesis. The p-value of 90% indicates that the actual page should be maintained.


Now, we need to conduct a z-test to compute the test statistic and p-value for evaluating if there is a statistically sigficance difference in conversion rates of the treatment group and the conversion rates of the control group.
converted_old,converted_new represent the sum of conversion in each group
n_old, n_new represent the size of each group.

```python
converted_old = control.converted.sum()
converted_new = treatment.converted.sum()

print("converted_old: ",converted_old)
print("converted_new: ",converted_new)
print("n_new: ",n_new)
print("n_old: ",n_old)
```

```python
z_score, p_value = sm.stats.proportions_ztest([converted_new,converted_old],[n_new, n_old],alternative='larger')
print("z_score: ",z_score)
print("p_value: ",p_value)
```
